# Python for Data Analytics
# Pandas (Data table analysis)

```python
import pandas as pd

data = {
    "Name": ["Rahim", "Karim", "Shibli"],
    "Sales": [100, 200, 150]
}

df = pd.DataFrame(data)

print(df)
```

### Total Sales

```python
print(df["Sales"].sum())
```

### Average Sales

```python
print(df["Sales"].mean())
```

### Highest Sales

```python
print(df["Sales"].max())
```

### Lowest Sales

```python
print(df["Sales"].min())
```

### Specific Person Data

```python
print(df[df["Name"] == "Shibli"])
```

### New Column 
### if bonus 10%

```python
df["Bonus"] = df["Sales"] * 0.10
print(df)
```

###  Practice Code

```python
import pandas as pd

data = {
    "Name": ["Rahim", "Karim", "Shibli"],
    "Sales": [100, 200, 150]
}

df = pd.DataFrame(data)

print(df)

print("Total Sales:", df["Sales"].sum())
print("Average Sales:", df["Sales"].mean())
print("Highest Sales:", df["Sales"].max())
print("Lowest Sales:", df["Sales"].min())

df["Bonus"] = df["Sales"] * 0.10

print(df)
```

### Homework 

```python
import pandas as pd 
data = {
    "Product":['Mobile', 'Laptop', 'Mouse'],
    "Price": [20000, 50000, 1000]
}
df = pd.DataFrame(data)
print(df)

print("Total Price:", df["Price"].sum())
print("Average Price:", df["Price"].mean())
print("Highest Price:", df["Price"].max())
print("Lowest Price:", df["Price"].min())

df["Discount"] = df["Price"] * 0.05

print(df)
```
