### CSV = comma separated file

```python
import pandas as pd

df = pd.read_csv("sales.csv")

print(df)
```

```python
print(df["Sales"].sum())
```

```python
print(df["Sales"].mean())
```

```python
print(df["Sales"].max())
```

```python
print(df["Sales"].min())
```

```python
print(df.isnull())
```

```python
print(df.isnull().sum())
```

```python
import pandas as pd

df = pd.read_csv("sales.csv")

print(df)

print("Total Sales:", df["Sales"].sum())
print("Average Sales:", df["Sales"].mean())
print("Highest Sales:", df["Sales"].max())
print("Lowest Sales:", df["Sales"].min())

print(df.isnull().sum())

df["Sales"] = df["Sales"].fillna(0)

print(df)
```
