### Total sales

```python
sales = [100, 200, 150, 300]
total_sales = sum(sales)
print(total_sales)
```

### Average Sales

```python
sales = [100, 200, 150, 300]
average_sales = sum(sales) / len(sales)
print(average_sales)
```

## Data Analysis

```python
sales = [100, 200, 150, 300]

print("Total sales:", sum(sales))
print("Average sales:", sum(sales) / len(sales))
print("Highest sales:", max(sales))
print("Lowest sales:", min(sales))
```
