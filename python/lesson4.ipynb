### *Import Library*

```python
import pandas as pd
import matplotlib.pyplot as plt
```


### Data Create

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Name": ["Rahim", "Karim", "Shibli"],
    "Sales": [100, 200, 150]
}

df = pd.DataFrame(data)

print(df)
```

### Bar Chart

```python
plt.bar(df["Name"], df["Sales"])

plt.title("Sales Report")
plt.xlabel("Name")
plt.ylabel("Sales")

plt.show()
```

### Line Chart

```python
plt.plot(df["Name"], df["Sales"])
plt.title("Sales Trend")
plt.show()
```

### Pie Chart

- 1.1f : ১ digit decimal পর্যন্ত float number
- % means percentage

```python
plt.pie(df["Sales"], labels=df["Name"], autopct="%1.2f%%")
plt.title("Sales Share")
plt.show()
```
