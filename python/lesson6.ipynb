# GroupBY
- Which dept. sales most?
- which category best?
- which employee perform best?


## Data create

```python
import pandas as pd

data = {
    "Product": ["Mobile", "Laptop", "Mobile", "Mouse", "Laptop"],
    "Sales": [200, 500, 300, 100, 200]
}

df = pd.DataFrame(data)

print(df)
```

```python
df.groupby("Product")["Sales"].sum()
```

```python
df.groupby("Product")["Sales"].mean()
```

```python
df.groupby("Product")["Sales"].max()
```

```python
df.groupby("Product")["Sales"].agg(["sum", "mean", "max", "min"])
```
