# Data Cleaning
- Dirty Data create

```python
import pandas as pd

data = {
    "Name": ["Rahim", "Karim", "Rahim", "  Shibli  "],
    "Sales": [100, None, 100, 200]
}

df = pd.DataFrame(data)

print(df)
```

- missing value check

```python
print(df.isnull())
```

- missing amount

```python
print(df.isnull().sum())
```

- Missing value fix

```python
df["Sales"] = df["Sales"].fillna(0)

print(df)
```

### Duplicates remove

```python
df = df.drop_duplicates()

print(df)
```

### Extra Space remove
-  "  Shibli  "
-  "Shibli"

```python
df["Name"] = df["Name"].str.strip()

print(df)
```

### Lowercase / Uppercase fix
- rahim
- RAHIM
- Rahim

```python
df["Name"] = df["Name"].str.lower()

print(df)
```

## Cleaning code

```python
import pandas as pd

data = {
    "Name": ["Rahim", "Karim", "Rahim", "  Shibli  "],
    "Sales": [100, None, 100, 200]
}

df = pd.DataFrame(data)

print("Original Data:")
print(df)

# Missing Value Fix
df["Sales"] = df["Sales"].fillna(0)

# Remove Duplicates
df = df.drop_duplicates()

# Remove Extra Spaces
df["Name"] = df["Name"].str.strip()

# Lowercase
df["Name"] = df["Name"].str.lower()

print("\nCleaned Data:")
print(df)
```
